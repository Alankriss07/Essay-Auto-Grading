```python
def pos_tags(data):
    sent = tokenize_sentences(data)[0]
    
    noun_count = 0
    adj_count = 0
    verb_count = 0
    adv_count = 0
    adp_count=0
    conj_count = 0
    det_count = 0
    prt_count = 0
    pron_count = 0
    punct_count=0
    
    
    for s in sent:
        tags = nltk.pos_tag(tokenize_words(s)[0])
        #print(tags)
        for tag in tags:
            if tag[1][0] == 'N':
                noun_count += 1
            elif tag[1][0] == 'J':
                adj_count += 1
            elif tag[1][0] == 'V':
                verb_count += 1
            elif tag[1][0] == 'R':
                adv_count += 1
            elif tag[1] == 'IN':
                adp_count += 1
            elif tag[1] == 'CC':
                conj_count += 1
            elif tag[1] == 'DT':
                det_count += 1
            elif tag[1] == 'PRT':
                prt_count += 1
            elif tag[1] == 'PRP' or tag[1] == 'PRP$' :
                pron_count += 1
            elif tag[1] == '.':
                punct_count += 1
                
    return noun_count,adj_count,verb_count,adv_count , adp_count , conj_count, det_count , pron_count , punct_count


pos_tags(" new good on at relly already and or the some year home at on he their his say told ! ")
```
